```python
# 1.Ans.
import seaborn as sns
import plotly.express as px

titanic_df = sns.load_dataset('titanic')
scatter_plot = px.scatter(titanic_df, x='age', y='fare')
scatter_plot.show()
```

```python
# 5.Ans.

Distplot is a type of plot that shows the distribution of a continuous variable. 
It can be used to visualize the shape, spread, and central tendency of a dataset.
```

```python
import plotly.express as px
import pandas as pd

tips_df = px.data.tips()

fig = px.histogram(tips_df, x="total_bill", nbins=30, opacity=0.7, color_discrete_sequence=['purple'])

fig.update_layout(
    title_text='Distribution of Total Bill',
    xaxis_title_text='Total Bill',
    yaxis_title_text='Count'
)

fig.show()
```
